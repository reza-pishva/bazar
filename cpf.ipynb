```python
import pandas as pd

# خواندن فایل CPF
cpf_df = pd.read_excel('CPF.xlsx')  # یا pd.read_csv('CPF.csv') اگر CSV باشه

# تمیز کردن نام ستون‌ها
cpf_df.columns = cpf_df.columns.str.strip()

# تغییر نام ستون‌ها برای هماهنگی با فایل دوم
cpf_df.rename(columns={'Date': 'DATE_SHAMSI', 'Hour': 'HOUR'}, inplace=True)

# خواندن فایل merged_output2.xlsx
merged_df = pd.read_csv('merged_output2.csv')
merged_df.columns = merged_df.columns.str.strip()

# تبدیل نوع داده‌ها به رشته برای تطابق دقیق
cpf_df['DATE_SHAMSI'] = cpf_df['DATE_SHAMSI'].astype(str)
cpf_df['HOUR'] = cpf_df['HOUR'].astype(str)
merged_df['DATE_SHAMSI'] = merged_df['DATE_SHAMSI'].astype(str)
merged_df['HOUR'] = merged_df['HOUR'].astype(str)

# ادغام دو دیتافریم بر اساس تاریخ و ساعت
merged_df = merged_df.merge(
    cpf_df[['DATE_SHAMSI', 'HOUR', 'CPF']],
    on=['DATE_SHAMSI', 'HOUR'],
    how='left'
)

# جایگزینی مقدار ستون importance_factor با CPF
merged_df['importance_factor'] = merged_df['CPF']
merged_df.drop(columns=['CPF'], inplace=True)

# ذخیره فایل نهایی
merged_df.to_excel('merged_output2_updated.xlsx', index=False)
```
