```python
import pandas as pd

# مسیر فایل‌ها
power_code_path = r"D:\BackUp\power_code_clean.csv"
ebraz_path = r"D:\BackUp\ebraz.csv"
output_path = r"D:\BackUp\actual_ebraz.csv"

# خواندن فایل‌ها
df_power = pd.read_csv(power_code_path)
df_ebraz = pd.read_csv(ebraz_path)

# ادغام: ID_EBRAZ از power_code با ID_NO از ebraz
df_merged = df_power.merge(
    df_ebraz[['ID_NO', 'POWER']], 
    left_on='ID_EBRAZ', 
    right_on='ID_NO', 
    how='left'
)

# تغییر نام ستون POWER از جدول دوم به EBRAZ
df_merged.rename(columns={'POWER_y': 'EBRAZ', 'POWER_x': 'POWER'}, inplace=True)

# انتخاب ستون‌های مورد نظر
final_columns = [
    'HOUR',
    'DATE_MILADI',
    'DATE_SHAMSI',
    'POWER',
    'CODE',
    'UNIT_NO',
    'DAMA',
    'ROTOOBAT',
    'ID_EBRAZ',
    'EBRAZ'
]

df_final = df_merged[final_columns]

# ذخیره در فایل خروجی
df_final.to_csv(output_path, index=False, encoding='utf-8-sig')

print("فایل actual_ebraz.csv با موفقیت ساخته شد ✅")
```
