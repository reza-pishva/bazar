```python
import pandas as pd

# ---------- تنظیم مسیر فایل‌ها ----------
input_excel_path = "merged_output2.csv"
output_excel_path = "merged_output2_with_importance2.xlsx"

# ---------- خواندن فایل اکسل ----------
df = pd.read_csv(input_excel_path)

# ---------- نگاشت نوع بار به ضریب ----------
factor_map = {
    "کم باری": 1.0,
    "میانباری": 1.5,
    "پرباری": 2.5
}

# ---------- جدول TOU سال 1403 ----------
# کلید: ماه شمسی (01 تا 12)
# مقدار: لیست 24 ساعته نوع بار
tou_1403 = {
    "01": ["کم باری"]*9 + ["میانباری"]*11 + ["پرباری"]*2 + ["میانباری"]*2,
    "02": ["کم باری"]*9 + ["میانباری"]*11 + ["پرباری"]*2 + ["میانباری"]*2,
    "03": ["میانباری","میانباری"] + ["کم باری"]*6 + ["میانباری"]*4 + ["پرباری"]*5 + ["میانباری"]*7,
    "04": ["میانباری","میانباری"] + ["کم باری"]*6 + ["میانباری"]*4 + ["پرباری"]*5 + ["میانباری"]*7,
    "05": ["میانباری","میانباری"] + ["کم باری"]*6 + ["میانباری"]*4 + ["پرباری"]*5 + ["میانباری"]*7,
    "06": ["میانباری","میانباری"] + ["کم باری"]*6 + ["میانباری"]*4 + ["پرباری"]*5 + ["میانباری"]*7,
    "07": ["کم باری"]*9 + ["میانباری"]*8 + ["پرباری"]*4 + ["میانباری"]*3,
    "08": ["کم باری"]*9 + ["میانباری"]*8 + ["پرباری"]*4 + ["میانباری"]*3,
    "09": ["کم باری"]*9 + ["میانباری"]*8 + ["پرباری"]*4 + ["میانباری"]*3,
    "10": ["کم باری"]*9 + ["میانباری"]*8 + ["پرباری"]*4 + ["میانباری"]*3,
    "11": ["کم باری"]*9 + ["میانباری"]*8 + ["پرباری"]*4 + ["میانباری"]*3,
    "12": ["کم باری"]*9 + ["میانباری"]*8 + ["پرباری"]*4 + ["میانباری"]*3,
}

# ---------- تابع محاسبه ضریب ----------
def calculate_importance_factor(row):
    try:
        # استخراج ماه از تاریخ شمسی (YYYY/MM/DD)
        month = row["DATE_SHAMSI"].split("/")[1]
        
        # تبدیل ساعت به اندیس (1-24 → 0-23)
        hour_index = int(row["HOUR"]) - 1
        
        load_type = tou_1403[month][hour_index]
        return factor_map[load_type]
    except:
        return None

# ---------- اضافه کردن ستون ----------
df["importance_factor"] = df.apply(calculate_importance_factor, axis=1)

# ---------- ذخیره خروجی ----------
df.to_excel(output_excel_path, index=False)

print("فایل خروجی با موفقیت ساخته شد:", output_excel_path)
```
